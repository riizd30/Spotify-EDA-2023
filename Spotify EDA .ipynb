# Exploratory Data Analysis

 ### Use Function of compute_distribution

```python
def compute_distribution(df1, column_name):
    distribution = df1[column_name].value_counts(normalize=True) * 100
    sorted_distribution = distribution.sort_values(ascending=False).reset_index()
    sorted_distribution.columns = [column_name.capitalize(), 'Percentage']
    return sorted_distribution
```

### Use Function of plot_pie_chart

```python
def plot_pie_chart(df, title, subplot_position):
    plt.subplot(2, 2, subplot_position)
    plt.pie(df['Percentage'], labels=df[df.columns[0]], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel", len(data)))
    plt.title(title)
```

## Average % of Features Music and Mode Distribution

```python
def visualize_data(df1):

    # List of parameters to compute mean values
    para_list = ['danceability_%', 'energy_%', 'valence_%', 'acousticness_%', 'liveness_%', 'speechiness_%', 'instrumentalness_%']

    # Compute mean values using a dictionary comprehension
    paras = {para.split('_')[0].capitalize(): df[para].mean() for para in para_list}

    # Plotting
    plt.figure(figsize=(15, 10))

    # Bar chart for parameters
    plt.subplot(2, 2, 1)
    sns.barplot(x=list(paras.keys()), y=list(paras.values()), palette="pastel")
    plt.title('Average % of Music Features')
    plt.ylabel('%')
    plt.xticks(rotation=45)

    # Pie charts
    plot_pie_chart(compute_distribution(df1, 'mode'), 'Mode Distribution', 2)
    
    plt.tight_layout()
    plt.show()


# Visualize the data
visualize_data(df)
```

According to the data, 69% are original, compared to 31% who are not. This indicates that artists are more likely to be unique individuals than to be modeled after other artists in terms of form or style.

There are several ways to interpret the word unique. One interpretation is that it suggests these artists are wholly unique and uninfluenced by other artists. Another interpretation is that it suggests these artists have incorporated elements from other artists into their own creations, creating something fresh and unique.
